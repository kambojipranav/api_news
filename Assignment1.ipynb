```python
# Step 1: Install required packages
!pip install requests newspaper3k nltk textblob pandas matplotlib seaborn

# Step 2: Import libraries
import requests
import json
import pandas as pd
from datetime import datetime
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from google.colab import userdata, files

# Step 3: Setup API Key (NewsAPI)
print("Setting up API key...")

# You need to set this secret in Colab first:
# Go to: Runtime → Secrets → Add key:
# - GOOGLE_NEWS_API_KEY: your_newsapi_key

# Get secret from Colab
try:
    GOOGLE_NEWS_API_KEY = userdata.get('03e5585ba155444ca30bbbd7519beafc')
    print("API key loaded successfully!")
except Exception as e:
    print(f"Error loading secret: {e}")
    print("Please set up GOOGLE_NEWS_API_KEY in Colab secrets first!")
    # For testing, you can set it directly (but better to use secrets)
    GOOGLE_NEWS_API_KEY = "03e5585ba155444ca30bbbd7519beafc"  # Replace with your actual key

# Step 4: Initialize NLTK
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

# Step 5: Create the News Collector Class
class NewsDataCollector:
    def __init__(self):
        self.google_news_api_key = GOOGLE_NEWS_API_KEY
        self.sia = SentimentIntensityAnalyzer()

    def get_news(self, query="artificial intelligence", language='en', max_results=20):
        """
        Fetch news from News API
        """
        try:
            url = "https://newsapi.org/v2/everything"
            params = {
                'q': query,
                'language': language,
                'sortBy': 'publishedAt',
                'apiKey': self.google_news_api_key,
                'pageSize': max_results
            }

            print(f"🔍 Searching for: {query}")
            response = requests.get(url, params=params)
            response.raise_for_status()

            news_data = response.json()
            return self.process_news_data(news_data)

        except requests.exceptions.RequestException as e:
            print(f"❌ Error fetching news: {e}")
            return None

    def process_news_data(self, news_data):
        """
        Process and analyze news data with sentiment analysis
        """
        processed_articles = []

        if news_data and 'articles' in news_data:
            print(f"📊 Found {len(news_data['articles'])} articles")

            for article in news_data['articles']:
                # Perform sentiment analysis
                content = f"{article.get('title', '')} {article.get('description', '')}"
                sentiment = self.analyze_sentiment(content)

                processed_article = {
                    'title': article.get('title', 'No title'),
                    'description': article.get('description', 'No description'),
                    'url': article.get('url', '#'),
                    'published_at': article.get('publishedAt', ''),
                    'source': article.get('source', {}).get('name', 'Unknown'),
                    'author': article.get('author', 'Unknown'),
                    'content': article.get('content', ''),
                    'sentiment_score': sentiment['compound'],
                    'sentiment_label': self.get_sentiment_label(sentiment['compound']),
                    'positive_score': sentiment['pos'],
                    'negative_score': sentiment['neg'],
                    'neutral_score': sentiment['neu']
                }
                processed_articles.append(processed_article)

        return processed_articles

    def analyze_sentiment(self, text):
        """Analyze sentiment using NLTK's VADER"""
        return self.sia.polarity_scores(text)

    def get_sentiment_label(self, score):
        """Convert sentiment score to label"""
        if score >= 0.05:
            return "Positive"
        elif score <= -0.05:
            return "Negative"
        else:
            return "Neutral"

    def save_to_json(self, articles, filename=None):
        """Save collected data to JSON file"""
        if filename is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = f"news_data_{timestamp}.json"

        try:
            with open(filename, 'w', encoding='utf-8') as f:
                json.dump(articles, f, indent=2, ensure_ascii=False)
            print(f"💾 Data saved to {filename}")
            return filename
        except Exception as e:
            print(f"❌ Error saving to file: {e}")
            return None

    def save_to_csv(self, articles, filename=None):
        """Save collected data to CSV file"""
        if filename is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = f"news_data_{timestamp}.csv"

        try:
            df = pd.DataFrame(articles)
            df.to_csv(filename, index=False, encoding='utf-8')
            print(f"💾 Data saved to {filename}")
            return filename
        except Exception as e:
            print(f"❌ Error saving to CSV: {e}")
            return None

# Step 6: Data Visualization Class
class NewsVisualizer:
    def __init__(self):
        plt.style.use('default')

    def plot_sentiment_distribution(self, articles):
        """Plot sentiment distribution pie chart"""
        if not articles:
            return

        df = pd.DataFrame(articles)
        sentiment_counts = df['sentiment_label'].value_counts()

        plt.figure(figsize=(10, 6))
        colors = ['#ff9999', '#66b3ff', '#99ff99']
        plt.pie(sentiment_counts.values, labels=sentiment_counts.index,
                autopct='%1.1f%%', colors=colors, startangle=90)
        plt.title('📊 Sentiment Distribution of News Articles')
        plt.axis('equal')
        plt.show()

    def plot_sentiment_scores(self, articles):
        """Plot sentiment scores bar chart"""
        if not articles:
            return

        df = pd.DataFrame(articles)

        plt.figure(figsize=(12, 6))
        plt.bar(range(len(df)), df['sentiment_score'],
                color=df['sentiment_score'].apply(
                    lambda x: 'green' if x > 0.05 else ('red' if x < -0.05 else 'gray')
                ))
        plt.title('📈 Sentiment Scores of Articles')
        plt.xlabel('Article Index')
        plt.ylabel('Sentiment Score')
        plt.axhline(y=0.05, color='blue', linestyle='--', alpha=0.7, label='Positive Threshold')
        plt.axhline(y=-0.05, color='orange', linestyle='--', alpha=0.7, label='Negative Threshold')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    def plot_source_distribution(self, articles):
        """Plot news source distribution"""
        if not articles:
            return

        df = pd.DataFrame(articles)
        source_counts = df['source'].value_counts().head(10)

        plt.figure(figsize=(12, 6))
        source_counts.plot(kind='bar', color='skyblue')
        plt.title('🏢 Top News Sources')
        plt.xlabel('News Source')
        plt.ylabel('Number of Articles')
        plt.xticks(rotation=45, ha='right')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    def create_wordcloud(self, articles):
        """Create word cloud from article titles"""
        if not articles:
            return

        text = ' '.join([article['title'] for article in articles if article['title'] != 'No title'])

        wordcloud = WordCloud(width=800, height=400,
                             background_color='white',
                             colormap='viridis',
                             max_words=100).generate(text)

        plt.figure(figsize=(15, 8))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('☁️ Word Cloud of News Headlines')
        plt.show()

# Step 7: Training Class
class SentimentTraining:
    def __init__(self):
        self.sia = SentimentIntensityAnalyzer()

    def train_examples(self):
        """Show sentiment analysis training examples"""
        examples = [
            "This AI technology is revolutionary and amazing!",
            "I'm very disappointed with the software update.",
            "The company reported stable growth this quarter.",
            "Terrible customer service experience.",
            "Outstanding results and excellent performance!",
            "The market showed mixed signals today.",
            "Innovative breakthrough in quantum computing",
            "Massive data breach affects millions of users",
            "Moderate progress in climate negotiations",
            "Exceptional customer satisfaction ratings"
        ]

        print("🧠 SENTIMENT ANALYSIS TRAINING EXAMPLES")
        print("=" * 60)

        results = []
        for example in examples:
            sentiment = self.sia.polarity_scores(example)
            label = self.get_sentiment_label(sentiment['compound'])
            results.append({
                'text': example,
                'score': sentiment['compound'],
                'label': label
            })

            print(f"\n📝 Text: {example}")
            print(f"   🏷️  Label: {label}")
            print(f"   📊 Score: {sentiment['compound']:.3f}")
            print(f"   ✅ Positive: {sentiment['pos']:.3f}, ❌ Negative: {sentiment['neg']:.3f}, ⚪ Neutral: {sentiment['neu']:.3f}")

        return results

    def get_sentiment_label(self, score):
        if score >= 0.05:
            return "✅ Positive"
        elif score <= -0.05:
            return "❌ Negative"
        else:
            return "⚪ Neutral"

# Step 8: Main Execution
print("🚀 Starting News Data Collection System")
print("=" * 50)

# Initialize classes
collector = NewsDataCollector()
visualizer = NewsVisualizer()
trainer = SentimentTraining()

# Fetch news data
print("\n📡 Fetching news data...")
articles = collector.get_news(
    query="artificial intelligence machine learning",
    max_results=15
)

if articles:
    # Display articles
    print(f"\n📰 COLLECTED {len(articles)} NEWS ARTICLES")
    print("=" * 50)

    for i, article in enumerate(articles, 1):
        print(f"\n{i}. {article['title']}")
        print(f"   🏷️  Sentiment: {article['sentiment_label']} ({article['sentiment_score']:.3f})")
        print(f"   📍 Source: {article['source']}")
        print(f"   📅 Published: {article['published_at'][:10]}")

    # Save data
    print("\n💾 Saving data...")
    json_file = collector.save_to_json(articles)
    csv_file = collector.save_to_csv(articles)

    # Download files
    if json_file:
        files.download(json_file)
    if csv_file:
        files.download(csv_file)

    # Visualizations
    print("\n📊 Generating visualizations...")
    visualizer.plot_sentiment_distribution(articles)
    visualizer.plot_sentiment_scores(articles)
    visualizer.plot_source_distribution(articles)
    visualizer.create_wordcloud(articles)

    # Summary statistics
    df = pd.DataFrame(articles)
    print(f"\n📈 SUMMARY STATISTICS")
    print("=" * 30)
    print(f"Total articles: {len(df)}")
    print(f"Positive articles: {len(df[df['sentiment_label'] == 'Positive'])}")
    print(f"Negative articles: {len(df[df['sentiment_label'] == 'Negative'])}")
    print(f"Neutral articles: {len(df[df['sentiment_label'] == 'Neutral'])}")
    print(f"Average sentiment score: {df['sentiment_score'].mean():.3f}")
    print(f"Most common source: {df['source'].mode().iloc[0] if not df['source'].mode().empty else 'N/A'}")

else:
    print("❌ No articles were fetched. Please check your API key.")

# Step 9: Training Section
print("\n" + "=" * 60)
print("🎓 SENTIMENT ANALYSIS TRAINING")
print("=" * 60)

training_results = trainer.train_examples()

# Step 10: Advanced Analysis (Optional)
if articles:
    print("\n" + "=" * 60)
    print("🔍 ADVANCED ANALYSIS")
    print("=" * 60)

    df = pd.DataFrame(articles)

    # Sentiment by source
    print("\n📊 Sentiment by News Source:")
    sentiment_by_source = df.groupby('source')['sentiment_score'].agg(['mean', 'count']).round(3)
    print(sentiment_by_source.sort_values('mean', ascending=False))

    # Time analysis (if dates are properly formatted)
    try:
        df['published_date'] = pd.to_datetime(df['published_at']).dt.date
        daily_sentiment = df.groupby('published_date')['sentiment_score'].mean()
        if len(daily_sentiment) > 1:
            print(f"\n📅 Daily Sentiment Trend:")
            print(daily_sentiment)
    except:
        pass

print("\n✅ News data collection and analysis completed!")
```
